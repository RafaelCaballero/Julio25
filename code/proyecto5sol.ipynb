```python
url = "https://raw.githubusercontent.com/tejal04/Machine-Learning-Practice/master/Data/TeleCustomers.csv"
import pandas as pd
df = pd.read_csv(url)
df
```

Se trata de datos de usuarios de un sistema de pago de televisión.

Metadatos:
    
    - region: región donde vive el usuario
    - tenure: meses que lleva en la compañía
    - age: edad
    - marital: 1 casado, 0 soltero
    - address: un indicador de la dirección (no explicado)
    - income: ingresos anuales en miles
    - ed: nivel educativo (valor más alto mayor nivel)
    - employ: años que ha trabajado en total
    - retire: 0 si está en activo, 1 si está jubilado
    - gender: 0 mujer, 1 hombre
    - reside: tipo de residencia: en propiedad, alquiler, etc. (no detallado)
    - cuscat: 1- Basic Service, 2- E-Service, 3- Plus Service 4- Total Service

**Ejercicio 3** Escribir código para determinar si el sueldo medio de las mujeres es superior al de los hombres o viceversa

```python
filtroMujer = df.gender==0
df[filtroMujer].income.mean(), df[~filtroMujer].income.mean()
```

```python
from scipy.stats import kstest
kstest(df[filtroMujer].income, df[~filtroMujer].income)
```

**Ejercicio 4** Gráfico de tarta de la columna `ed`

```python
import matplotlib.pyplot as plt
%matplotlib inline

frecuencias =  df.ed.value_counts()
labels = frecuencias.index
sizes = frecuencias.values

plt.pie(sizes, labels=labels)
plt.show()
```
